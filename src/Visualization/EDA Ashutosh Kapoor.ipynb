```
import glob
import pandas as pd
path = "project"
f = glob.glob(path+"/*.csv")
lst = []
for i in f:
    df = pd.read_csv(i, index_col = None, header=0)
    lst.append(df)
frame = pd.concat(lst,axis=0,ignore_index=True)
frame.rename(columns=({"0":"date","1":"data"}),inplace=True)
print(frame.to_string())
```

```
from lets_plot import * 
ggplot() + \
geom_bar(aes(x="date", y="data"), data=frame, sampling="none" if frame.size < 50 else sampling_pick(n=50), stat="identity") + \
ggtitle("Chart") 
```

```
from lets_plot import * 
ggplot() + \
geom_area(aes(x="date", y="data"), data=frame, sampling="none" if frame.size < 2500 else sampling_systematic(n=2500), stat="identity", position="identity") + \
ggtitle("Chart") 
```

```
from lets_plot import * 
ggplot() + \
geom_smooth(aes(x="date", y="data"), data=frame, sampling="none" if frame.size < 2500 else sampling_systematic(n=2500), color="#ff00d0") + \
ggtitle("Chart") 
```

```
from lets_plot import * 
ggplot()
from lets_plot.bistro import corr
corr.corr_plot(frame).points().build() + ggsize(500, 370) + \
ggtitle("Chart") 
```
